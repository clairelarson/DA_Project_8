```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme()
palette = ['#292F36','#33575A','#3C7E7D','#4ECDC4','#A3E6DE','#F7FFF7','#FF6B6B','#FF8A6C','#FFA96C','#FDA53A','#FFE66D','#CFC664','#a1c181','#619b8a','#3D5228']

movies = pd.read_csv('movies.csv')
movies.head()
```

# Questions:

## What was the runtime for each movie?

```python
runtime = sns.catplot(data=movies, x='Movie Title', y='Runtime', kind='bar', aspect=3, palette = palette)
plt.xticks(rotation=85, fontsize=20)
plt.title('Runtime for Each Movie', fontsize = 20)
plt.xlabel("Movie")
plt.ylabel("Runtime")
plt.show()
runtime.savefig("runtime.png")
```

## Notes:
The runtime for each movie was between 100 to 160 minutes. Most movies fell between 120 to 140 minutes.

## What movie had the longest runtime?

```python
long = movies[movies['Movie Title']=='Dune']
long
```

## Notes:
As can be seen from the graph above, Dune had the longest runtime at 155 minutes. 

## Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?

```python
sns.regplot(data=movies, x='Runtime', y='Award Wins', color="#de5a79")
plt.title('Relationship Between Runtime and the Number of Awards', fontsize = 20)
plt.xlabel("Runtime")
plt.ylabel("Awards Won")
plt.show()
runtime.savefig("runtime.png")
```

## Notes:
There is a slight positive relationship between runtime and awards won. However, this could be due to Dune, which is an outlier for runtime.

## (Graph) How many awards and nominations did each movie receive?

```python
noms = sns.catplot(data=movies, x='Movie Title', y='Award Nominations', kind='bar', aspect=3, palette = palette)
plt.xticks(rotation=85, fontsize=20)
plt.title('Award Nominations for Each Movie', fontsize = 20)
plt.xlabel("Movie")
plt.ylabel("Awards Nominated")
plt.show()
noms.savefig("noms.png")
```

```python
wins = sns.catplot(data=movies, x='Movie Title', y='Award Wins', kind='bar', aspect=3, palette = palette)
plt.xticks(rotation=85, fontsize=20)
plt.title('Award Wins for Each Movie', fontsize = 20)
plt.xlabel("Movie")
plt.ylabel("Awards Won")
plt.show()
wins.savefig("wins.png")
```

## Notes:
Parasite won the most awards, but The Shape of Water was nominated for the most awards. Free Solo was both the least nominated and the least awarded movie out of the 15 movies selected.

# Extra Credit:

## Is there a relationship between the amount of box office earnings a movie had and the amount of total nominations the movie received (total nominations = awards wins + award nominations)?

```python
total_noms = movies['Award Wins'] + movies['Award Nominations']
movies['Total Nominations'] = total_noms
sns.regplot(data=movies, x='Box Office', y='Total Nominations', color="#4ECDC4")
plt.title('Box Office Earnings and Total Nominations', fontsize = 20)
plt.xlabel("Box Office")
plt.ylabel("Total Noms")
plt.show()
```

## Notes:
There is a weak negative relationship between box office earnings and total nominations. The relationship is too weak to hold significance.

## Is there a relationship between box office earnings and movie runtimes?

```python
sns.regplot(data=movies, x='Box Office', y='Runtime', color="#3D5228")
plt.title('Box Office Earnings and Runtime', fontsize = 20)
plt.xlabel("Box Office")
plt.ylabel("Runtime")
plt.show()
```

## Notes:
There is no strong correlation between box office earnings and the run time of the movie.

## (Graph) How much did each movie earn at the box office?

```python
box = sns.catplot(data=movies, x='Movie Title', y='Box Office', kind='bar', aspect=3, palette = palette)
plt.xticks(rotation=85, fontsize=20)
plt.title('Box Office Earnings for Each Movie', fontsize = 20)
plt.xlabel("Movie")
plt.ylabel("Box Office")
plt.show()
box.savefig("box.png")
```
